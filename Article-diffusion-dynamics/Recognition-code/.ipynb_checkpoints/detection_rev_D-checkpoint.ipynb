```python
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import typing as tp

from matplotlib import pyplot as plt

sns.set()
```

```python
def make_dataset(frames):
    """
    Creates snapshots.
    """
    frame = frames[100]
    for x in range(0, len(frame) - 50, 5):
        for y in range(0, len(frame) - 50, 5):
            cv2.imwrite('D:/Documents/Creaturae/Active/Video/train/'
                        + str(100) + '_' + str(x) + '_' + str(y) + '.jpg', frame[y:y + 50, x: x + 50])
```

```python
def extract_frames(path: str, resize: tp.Optional[int] = None) -> tp.List[tp.Any]:
    """
    Extracts frames from video and resizes them, returns the list of images.
    """
    
    result: tp.List[tp.Any] = []
    vc = cv2.VideoCapture(path)
    success = True
    
    while success:
        success, image = vc.read()
        if success:
            if resize:
                 image = cv2.resize(image, (resize, resize))
            result.append(image)
            
    return result
```

```python
def get_circles(frames: tp.List[tp.Any], stop: int, start: int = 0) -> tp.List[tp.Any]:
    """
    Detects circles on image, returns the sequence of (x, y, radius).
    """
    
    result = []
    
    for i in range(start, stop):
        blur = cv2.medianBlur(cv2.cvtColor(frames[i], cv2.COLOR_BGR2GRAY), 5)
        circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1.28, 30, param1=50, param2=30, minRadius=18, maxRadius=32)
        result.append(circles[0])
        
    return result
```

```python
def find_trajectory(circles: tp.List[tp.List[tp.List[float]]]) -> tp.List[tp.List[tp.List[float]]]:
    """
    Finds trajectory of every circle detected on the first frame of sequence.
    Returns lists of (x, y) positions for every circle.
    """
    
    def dist(lhs: tp.List[float], rhs: tp.List[float]) -> float:
        return np.sqrt((lhs[0] - rhs[0]) ** 2 + (lhs[1] - rhs[1]) ** 2)
    
    def find_position(circle: tp.List[float], options: tp.List[tp.List[float]], max_dist: float = 30) -> tp.List[float]:
        """
        Finds the next position of the circle. If no circles near current were detected returns current position.
        """
        
        most_common = None
        for option in options:
            distance = dist(circle, option)
            if distance < max_dist and (most_common is None or dist(circle, most_common) > distance):
                most_common = option
                
        if most_common is None:
            most_common = circle
        return most_common
    
    result = [circles[0]]
    for i in range(len(circles) - 1):
        previous = result[-1]
        current = [find_position(previous[j], circles[i + 1]) for j in range(len(previous))]
        result.append(current)
        
    trajectory = []
    for i in range(len(result[0])):
        x, y = [], []
        for frame in range(len(circles)):
            x.append(result[frame][i][0])
            y.append(result[frame][i][1])
        trajectory.append([x, y])
        
    return trajectory
```

```python
def save_csv(trajectories: tp.List[tp.List[float]], path: str) -> None:
    """
    Saves list of trajectories to .csv.
    """
    data = {str(i) + '_x': trajectories[i][0] for i in range(len(trajectories))}
    data.update({str(i) + '_y': trajectories[i][1] for i in range(len(trajectories))})
    pd.DataFrame(data={key: data[key] for key in sorted(data)}).to_csv(path)
```

```python
def get_trajectories(path: str, stop: int = 1000) -> None:
    """
    Get trajectories of robots for the video.
    """
    frames = extract_frames(path)    
    circles = get_circles(frames, stop=stop)
    trajectories = find_trajectory(circles)
    return trajectories
```

```python
def plot_trajectories(trajectories: tp.List[tp.List[float]], number: int=-1) -> None:
    """
    Plots the trajectories.
    """
    for trajectory in trajectories[:number]:
        plt.plot(trajectory[0], trajectory[1])
```

```python
def calculate_length(trajectory: tp.List[float], stop=-1) -> None:
    """
    Finds the path length of trajectory.
    """
    
    def dist(i: int, j: int) -> float:
        return np.sqrt((trajectory[0][i] - trajectory[0][j]) ** 2 + (trajectory[1][i] - trajectory[1][j]) ** 2)
    
    current = 0
    length = [0]
    for i in range(1, len(trajectory[0][:stop])):
        current += dist(i - 1, i)
        length.append(current)
        
    return length
```

```python
def calculate_distance(trajectory: tp.List[float], stop=-1) -> None:
    """
    Finds the robot distance from the start point.
    """
    
    def dist(i: int, j: int) -> float:
        return np.sqrt((trajectory[0][i] - trajectory[0][j]) ** 2 + (trajectory[1][i] - trajectory[1][j]) ** 2)
    
    distance = [0]
    for i in range(1, len(trajectory[0][:stop])):
        distance.append(dist(0, i))
        
    return distance
```

```python
def calculate_median(arrays: tp.List[tp.List[float]]) -> None:
    """
    Finds the median of lists.
    """
    result = []
    for i in range(len(arrays[0])):
        median = 0
        for arr in arrays:
            median += arr[i]
        result.append(median)
        
    return result
```

```python
def plot_distance(array: tp.List[float], step: float = 1 / 15, log = False):
    """
    Plots array values.
    """
    if log:
        plt.loglog([i * step for i in range(len(array))], array)
    else:
        plt.plot([i * step for i in range(len(array))], array)
```

```python
def plot_median(trajectories: tp.List[tp.List[tp.List[float]]], log = False):
    """
    Calculates the median distance and lengts and plots them.
    """
    
    distance, length = [], []
    for trajectory in trajectories:
        distance.append(calculate_distance(trajectory))
        length.append(calculate_length(trajectory))
        
    plt.figure(figsize=(10, 10))
    plt.subplot(2, 1, 1)
    plot_distance(calculate_median(distance), log=log)
    plt.subplot(2, 1, 2)
    plot_distance(calculate_median(length), log=log)
    plt.show()
```
